 Q1. How do you comment code in Python? What are the different types of comments?

Ans:- In Python, comments are used to explain code and make it easier to understand. The interpreter ignores comments when running the program.

1. Single-line comments:-
Single-line comments start with the # symbol.

2. Multi-line comments (using #):-
Python does not have a special syntax for multi-line comments, but you can use multiple # symbols.

3. Docstrings (documentation comments):-
Docstrings are written using triple quotes (""" """ or ''' ''').
They are mainly used to describe modules, functions, classes, or methods and can be accessed at runtime using __doc__.

```python
#1.Single-line comments:-

x=19 #value store in x var and value is 19.
```

```python
#2. Multi-line comments (using #):-

# x=89
# y=78
# sum=x+y
# print(sum)
```

 Q4. How do you write and execute a Python script from the command line?

Ans:- 1. Write a Python Script

Open a text editor (Notepad, VS Code, nano, etc.).

Write your Python code.

Example:

print("Hello, World!")


Save the file with a .py extension, for example:

hello.py

2. Open the Command Line

Windows: Open Command Prompt or PowerShell

macOS/Linux: Open Terminal

3. Navigate to the Script Location

Use the cd command to go to the folder where the script is saved:

cd path_to_your_file


Example:

cd Desktop

4. Execute the Python Script
On Windows:
python hello.py


or

py hello.py

On macOS / Linux:
python3 hello.py

5. Output

The result will be displayed in the terminal:

Hello, World!
