```python
def test_fib(n):
    a,b=0,1
    for i in range(n):
        yield a
        a,b=b,a+b
```

```python
test_fib(9)
```

```python
for i in test_fib(10):
    print(i)
```

```python
fib=test_fib()
```

```python
def count_test(n):
    count=1
    while count<=n:
        yield count
        count +=1
```

```python
C=count_test(6)
```
