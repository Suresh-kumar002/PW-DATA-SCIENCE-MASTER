 Q3. Explain the use of the following operators using an example:
 (i)	/
 (ii)	%

```python
#(i) /:-Division
#(ii) % :- Modulus
#(iii) //:-Floor Division
#(iv) **:-Exponentiation

a=10
b=3

Division=a/b
Modulus=a%b
Floor_Division=a//b
Exponentiation=a**b

print(Division)
print(Modulus)
print(Floor_Division)
print(Exponentiation)
```

Q5. Using a while loop, verify if the number A is purely divisible by number B and if so then how many 
times it can be divisible.

```python
a=int(input("Enter a="))
b=int(input("Enter b="))
divisible=0

while a%b==0:
    a=a//b
    divisible +=1
if divisible > 0:
    print("Purely divisible", divisible, "times")
else:
    print("Not purely divisible")
```
