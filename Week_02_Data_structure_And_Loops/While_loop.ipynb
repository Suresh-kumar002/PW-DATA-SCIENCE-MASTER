#Finonacci series

```python
num=int(input("Enter the number of element you are looking for: "))
a=0
b=1
counter =0

while counter<num:
    print(a)
    c=a+b
    a=b
    b=c
    counter +=1
```

#print tables of number

```python
n=int(input("Enter the number: "))
i=1

while i<=10:
    result=n*i
    print(f"{n}x{i}={result}")
    i+=1
```

```python
n=int(input("Enter the number: "))
i=1

for i in range(1,11):
    result=n*i
    print(f"{n}x{i}={result}")
```
